```python
r = requests.get(master_url)
```

```python
playlists_url = m3u8.loads(r.text).data['playlists']
```

```python
playlists_url
```

```python
for playlist in playlists_url:
  stream_info = playlist['stream_info']
  if stream_info['resolution'] == '1920x1080':
    p1080_link = playlist['uri']
    break


```

```python
print(p1080_link)
```

```python
r = requests.get(p1080_link)
```

```python
url_info = m3u8.loads(r.text).data
```

```python
segments_info = url_info['segments']
```

```python
segments_info
```

```python
with open('video.ts','wb') as f:
  for segment in segments_info:
    ts_content = requests.get(segment['uri']).content
    f.write(ts_content)


```
